# YOLOv8n model average Inference time and FPS calculation


```python
import time
from ultralytics import YOLO

model = YOLO('/content/gdrive/MyDrive/yolov8n-wood/best-YOLOv8n.pt')
source = '/content/gdrive/MyDrive/yolov8n-wood/test'

start_time = time.time()
results = model.predict(source=source, save=True)
end_time = time.time()

fps = len(results) / (end_time - start_time)
print(f"FPS: {fps}")
```
